```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

filename = "results.csv"
```

```python
df_ttc = pd.read_csv(filename, sep=";")
df_ttc.columns
```

```python
df_ttc.columns
```

### Consider only the first 30 images

```python
df_ttc_30 = df_ttc[df_ttc['Image'].between(1,30)]
```

```python
df_ttc_30_gr = df_ttc_30.groupby(by=['Image'])[['TTC_LiDAR', 'TTC_Cam',]].mean()
df_ttc_30_gr.plot(title="TTC for Images 1-30 avg. over all Detector Combinations")
```

```python
df_ttc_30_gr = df_ttc_30.groupby(by=['Detector','Descriptor'])[['delta_TTC']].mean()
```

```python
df_ttc_30_gr.plot(kind='bar', title='Mean TTC differnce over 30 images bewteen LiDAR and Cams by Detector/Descriptor combination')
```
